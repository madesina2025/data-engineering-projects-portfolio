```python
# Step 1: Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, roc_auc_score, roc_curve
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import xgboost as xgb
import joblib

from imblearn.over_sampling import SMOTE

import warnings
warnings.filterwarnings('ignore')
```

```python
# Step 2: Load and Deduplicate Dataset
df = pd.read_csv('data/creditcard.csv')
print("Original shape:", df.shape)

# Drop duplicates only from non-fraud data
fraud_df = df[df['Class'] == 1]
non_fraud_df = df[df['Class'] == 0].drop_duplicates()
df = pd.concat([fraud_df, non_fraud_df], ignore_index=True)

print("After deduplication:", df.shape)
```

```python
# Step 3: Standardize and Apply SMOTE
scaler = StandardScaler()
df[['Amount', 'Time']] = scaler.fit_transform(df[['Amount', 'Time']])

X = df.drop('Class', axis=1)
y = df['Class']

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)
print("After SMOTE:", X_res.shape, y_res.shape)
```

```python
# Step 4: Split Data
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)
```

```python
# Step 5: Define Models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "Naive Bayes": GaussianNB(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Support Vector Machine": SVC(probability=True),
    "XGBoost": xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}
```

```python
# Step 6: Evaluate Models
model_results = []
model_store = {}

for name, model in models.items():
    print(f"Training: {name}")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    model_results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1 Score": f1_score(y_test, y_pred),
        "AUC": roc_auc_score(y_test, y_prob)
    })

    model_store[name] = model
```

```python
import joblib
import os

# Ensure models folder exists
os.makedirs("models", exist_ok=True)

# Save each trained model
for name, model in model_store.items():
    filename = f"models/{name.lower().replace(' ', '_')}_fraud_model.pkl"
    joblib.dump(model, filename)
    print(f"✅ Saved: {filename}")
```

```python
# Step 7
# Step 7: Save Comparison & Plot
import pandas as pd
import matplotlib.pyplot as plt

# Load the model results
results_df = pd.DataFrame(model_results)

# Save to CSV
results_df.to_csv("data/model_comparison.csv", index=False)

# Convert metrics to percentage for display
metrics = ["Accuracy", "Precision", "Recall", "F1 Score", "AUC"]
results_df[metrics] = results_df[metrics] * 100

# Plot
ax = results_df.set_index("Model")[metrics].plot.bar(figsize=(12, 6))

# Add percentage labels to bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='edge', fontsize=9, padding=3)

# Final touches
plt.title("Model Performance Comparison")
plt.ylabel("Score (%)")
plt.ylim(0, 110)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load model results
results_df = pd.read_csv("data/model_comparison.csv")

# Define metrics to visualize
metrics = ["Accuracy", "AUC", "Recall", "F1 Score", "Precision"]

# Assign a color to each model
model_colors = {
    "Logistic Regression": "#1f77b4",  # blue
    "Random Forest": "#ff7f0e",        # orange
    "Naive Bayes": "#2ca02c",          # green
    "K-Nearest Neighbors": "#d62728",  # red
    "Support Vector Machine": "#9467bd",  # purple
    "XGBoost": "#8c564b"               # brown
}

# Create one horizontal bar chart per metric
for metric in metrics:
    sorted_df = results_df.sort_values(by=metric, ascending=False)

    colors = [model_colors.get(model, "#333333") for model in sorted_df["Model"]]

    plt.figure(figsize=(10, 6))
    bars = plt.barh(sorted_df["Model"], sorted_df[metric], color=colors)
    plt.xlabel(f"{metric} (%)")
    plt.title(f"{metric} Comparison Across Models")
    plt.xlim(0, 1)
    plt.grid(axis="x")

    # Add values to bars
    for bar in bars:
        width = bar.get_width()
        plt.text(width + 0.01, bar.get_y() + bar.get_height()/2,
                 f"{width:.2%}", va='center')

    # Add custom legend
    handles = [plt.Rectangle((0,0),1,1, color=color) for color in model_colors.values()]
    labels = list(model_colors.keys())
    plt.legend(handles, labels, title="Models", loc="lower right")

    plt.tight_layout()
    plt.gca().invert_yaxis()
    plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import os

# Step 1: Load dataset
df = pd.read_csv("data/creditcard.csv")

# Step 2: Feature Scaling
scaler = StandardScaler()
df[['Amount', 'Time']] = scaler.fit_transform(df[['Amount', 'Time']])

# Step 3: Separate features and label
X = df.drop("Class", axis=1)
y = df["Class"]

# Create data directory if it doesn't exist
os.makedirs("data", exist_ok=True)

# Step 4: Save original imbalanced split
X_train_orig, X_test_orig, y_train_orig, y_test_orig = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

X_train_orig.to_csv("data/X_train_original.csv", index=False)
pd.Series(y_train_orig).to_csv("data/y_train_original.csv", index=False)

X_test_orig.to_csv("data/X_test_original.csv", index=False)
pd.Series(y_test_orig).to_csv("data/y_test_original.csv", index=False)

print("✅ Original (imbalanced) train-test split saved.")

# Step 5: Apply SMOTE to balance classes
sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X, y)

# Step 6: Resampled train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.3, random_state=42, stratify=y_resampled
)

X_train.to_csv("data/X_train.csv", index=False)
pd.Series(y_train).to_csv("data/y_train.csv", index=False)

X_test.to_csv("data/X_test.csv", index=False)
pd.Series(y_test).to_csv("data/y_test.csv", index=False)

print("✅ SMOTE-resampled train-test split saved.")
```

```python
import os
import joblib
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# --- Step 1: Setup ---
model_dir = "models_all"
output_dir = "reports/confusion_matrices"
os.makedirs(output_dir, exist_ok=True)

# Load test data
X_test = pd.read_csv("data/X_test.csv")
y_test = pd.read_csv("data/y_test.csv").values.ravel()

# List of saved models
model_files = [f for f in os.listdir(model_dir) if f.endswith(".pkl")]

# For storing metrics
conf_matrix_summary = []

# --- Step 2: Loop through models ---
for model_file in model_files:
    model_path = os.path.join(model_dir, model_file)
    model_name = model_file.replace("_fraud_model.pkl", "").replace("_", " ").title()

    # Load model and predict
    model = joblib.load(model_path)
    y_pred = model.predict(X_test)

    # Confusion matrix and metrics
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    acc = (tp + tn) / (tp + tn + fp + fn)
    prec = tp / (tp + fp) if (tp + fp) > 0 else 0
    rec = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1 = 2 * tp / (2 * tp + fp + fn) if (2 * tp + fp + fn) > 0 else 0

    # Append metrics summary
    conf_matrix_summary.append({
        "Model": model_name,
        "True Negative": tn,
        "False Positive": fp,
        "False Negative": fn,
        "True Positive": tp,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1 Score": f1
    })

    # --- Step 3: Plot Confusion Matrix ---
    cm = [[tn, fp], [fn, tp]]
    labels = ["Actual No", "Actual Yes"]
    cm_labels = [["TN", "FP"], ["FN", "TP"]]

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", cbar=True,
                xticklabels=labels, yticklabels=labels, linewidths=1, linecolor='white')

    for i in range(2):
        for j in range(2):
            plt.text(j + 0.1, i + 0.5, cm_labels[i][j], color="red", fontsize=14)

    plt.title(f"Confusion Matrix Heatmap for {model_name}")
    plt.ylabel("True Label")
    plt.xlabel("Predicted Label")
    plt.tight_layout()

    # Save plot
    save_path = os.path.join(output_dir, f"{model_name.lower().replace(' ', '_')}_conf_matrix.png")
    plt.savefig(save_path)
    plt.close()
    print(f"📊 Saved confusion matrix for: {model_name} → {save_path}")

# --- Step 4: Save Summary CSV ---
summary_df = pd.DataFrame(conf_matrix_summary)
summary_csv_path = "data/confusion_matrix_summary.csv"
summary_df.to_csv(summary_csv_path, index=False)
print(f"✅ Confusion matrix summary saved to: {summary_csv_path}")
```
