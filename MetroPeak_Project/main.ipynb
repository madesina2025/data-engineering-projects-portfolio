```python
import pandas as pd

def show_first_last(df, n_rows=5, n_cols=5):
    # choose columns: first n + last n (without duplicating if few columns)
    if df.shape[1] > 2*n_cols:
        cols = list(df.columns[:n_cols]) + list(df.columns[-n_cols:])
    else:
        cols = list(df.columns)

    head_part = df.loc[:, cols].head(n_rows).assign(_part="head")
    tail_part = df.loc[:, cols].tail(n_rows).assign(_part="tail")
    out = pd.concat([head_part, tail_part], axis=0)
    return out

# use it
display(show_first_last(df, n_rows=2, n_cols=2))
```

```python

# Focus on history.* and coerce common empties to NA
hist = (df.filter(regex=r'^history\.')
          .replace(['', 'NA', 'N/A', 'null', 'None'], pd.NA))

# Keep only well-formed triplets: history.YYYY-MM-DD.(event|date|price)
parts = hist.columns.to_series().str.extract(r'^history\.(\d{4}-\d{2}-\d{2})\.(event|date|price)$')
mask_valid = parts.notna().all(axis=1)
hist = hist.loc[:, mask_valid]
parts = parts[mask_valid].rename(columns={0: "date", 1: "field"})

# Build a MultiIndex on columns: (date, field)
hist.columns = pd.MultiIndex.from_arrays([parts["date"].values, parts["field"].values],
                                         names=["date","field"])

# Count non-nulls per date & field
summary_by_date = pd.DataFrame({
    "event_non_null": hist.xs("event", level="field", axis=1).notna().sum()
                        if ("event" in hist.columns.get_level_values("field")) else 0,
    "date_non_null":  hist.xs("date",  level="field", axis=1).notna().sum()
                        if ("date"  in hist.columns.get_level_values("field")) else 0,
    "price_non_null": hist.xs("price", level="field", axis=1).notna().sum()
                        if ("price" in hist.columns.get_level_values("field")) else 0,
})

summary_by_date["any_non_null"] = summary_by_date[["event_non_null","date_non_null","price_non_null"]].sum(axis=1) > 0

print("Dates with ANY value in the triplet:", int(summary_by_date["any_non_null"].sum()))
print("Dates with ALL-NULL triplets:", int((~summary_by_date["any_non_null"]).sum()))

# --- Inspect before dropping ---
# Dates where all three (event/date/price) are completely null
all_null_dates = summary_by_date.index[~summary_by_date["any_non_null"]].tolist()
summary_by_date.loc[all_null_dates].head(20)  # visual check

# (Optional) Peek at a few 'kept' dates
# summary_by_date.loc[summary_by_date["any_non_null"]].head(10)
```
